```python
def calculate_BMI(weight, height):
  BMI = weight / (height ** 2)
  if BMI <= 18.5:
    print("此BMI值属于偏瘦范围。")
  elif 18.5 < BMI <= 24.9:
    print("此BMI值属于正常范围。")
  elif 24.9 < BMI <= 29.9:
    print("此BMI值属于偏胖范围。")
  else:
    print("此BMI值属于肥胖范围。")
  return BMI
calculate_BMI(70, 1.76)
```

```python
class Student:
  def __init__(self, name, student_id):
    self.name = name
    self.student_id = student_id
    self.grades = {"chinese":0, "math":0, "english":0}

  def set_grade(self, course, grade):
    if course in self.grades:
      self.grades[course] = grade

  def print_grades(self):
    print(f"学生{self.name}(学号：{self.student_id}）的成绩为：")
    for course in self.grades:
      print(f"{course}:{self.grades[course]}")

chen = Student("xiaoming", 123456)
print(chen.name)
chen.set_grade("chinese", 90)
print(chen.grades)
chen.print_grades()
```

```python
class Employee:
  def __init__(self, name, id):
      self.name = name
      self.id = id
  def print_info(self):
    print(f"Name: {self.name},ID: {self.id}")

class FullTimeEmployee(Employee):
  def __init__(self, name, id, monthly_salary):
    super().__init__(name, id)
    self.monthly_salary = monthly_salary
  def calsuate_monthly_salary(self):
    return self.monthly_salary

class HourlyEmployee(Employee):
  def __init__(self, name, id, hourly_rate, hours_worked):
    super().__init__(name, id)
    self.hourly_rate = hourly_rate
    self.hours_worked = hours_worked
  def calculate_monthly_salary(self):
    return self.hourly_rate * self.hours_worked

xiaoming = FullTimeEmployee("xiaoming", 123456, 5000)
xiaohong = HourlyEmployee("xiaohong", 654321, 20, 160)

xiaoming.print_info()
print(xiaoming.calsuate_monthly_salary())

xiaohong.print_info()
print(xiaohong.calculate_monthly_salary())
```
